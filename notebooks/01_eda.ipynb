```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
current_dir = os.getcwd()
file_path = os.path.join(current_dir, "/Users/connor/insurance-expense-prediction-mlops", "data", "raw", "insurance.csv")
df = pd.read_csv(file_path)
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['charges'], kde=True)
plt.title('Sigorta Masrafları Dağılımı')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='smoker', y='charges', data=df)
plt.title('Sigara İçme Durumuna Göre Masraflar')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df)
plt.title('BMI vs Charges (Sigara kırılımı ile)')
plt.show()
```
